```python
import requests
from api import api_key
import pprint
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv("..\Outputs\movies.csv")
top_10_movies_popularity= movies.nlargest(10, 'Popularity')
top_10_movies_popularity
```

```python
top_10_movies_votes= movies.nlargest(10, 'Vote_Count')
top_10_movies_votes
```

```python
language_df = pd.read_csv("..\Outputs\Languages.csv")
language_df
```

```python
top_10_english_movies = movies[movies['Language_Id'] == 'en'].nlargest(10, 'Vote_Count')
top_10_english_movies
```

```python
top_10_foreign_movies = movies[movies['Language_Id'] != 'en'].nlargest(10, 'Vote_Count')
top_10_foreign_movies
```

```python
df=movies[movies['Language_Id']=='en']
df['Popularity'].sum()
```

```python
df2=movies[movies['Language_Id']=='es']
df2['Popularity'].sum()
```

```python
foriegn_movies = movies[movies['Language_Id'] != 'en']
fmovies_popcount = foriegn_movies.groupby('Language_Id')['Popularity'].sum().reset_index()
fmovies_popcount
```

```python
f, ax = plt.subplots(2, 2, figsize=(20, 20))

# Bar chart for 'Top 10 Movies by Popularity'
sns.barplot(x='Popularity', y='Title', data=top_10_movies_popularity,ax= ax[0,0], palette='Blues')
ax[0,0].set_xlabel('Popularity')
ax[0,0].set_ylabel('Movies')
ax[0,0].set_title('Top 10 Movies by Popularity')
ax[0,0].invert_yaxis()

# Bar chart for 'Top 10 Movies by Vote Count'
sns.barplot(x='Vote_Count', y='Title', data=top_10_movies_votes,ax= ax[0,1], palette='YlOrBr')
ax[0,1].set_xlabel('Vote Count')
ax[0,1].set_ylabel('Movies')
ax[0,1].set_title('Top 10 Movies by Vote Count')
ax[0,1].invert_yaxis()

# Bar chart for 'Top 10 English Language movies'
sns.barplot(x='Vote_Count', y='Title', data=top_10_english_movies,ax= ax[1,0], palette='pastel')
ax[1,0].set_xlabel('Vote Count')
ax[1,0].set_ylabel('Movies')
ax[1,0].set_title('Top 10 English Language Movies by Vote Count')
ax[1,0].invert_yaxis()

# Bar chart for 'Top 10 Foreign Language movies'
sns.barplot(x='Vote_Count', y='Title', data=top_10_foreign_movies,ax= ax[1,1], palette='pastel')
ax[1,1].set_xlabel('Vote Count')
ax[1,1].set_ylabel('Movies')
ax[1,1].set_title('Top 10 Foreign Language Movies by Vote Count')
ax[1,1].invert_yaxis()



plt.tight_layout()
plt.show()
```

```python
f, ax = plt.subplots(1, 1, figsize=(10, 6))

# Bar chart for 'Popularity of Foreign Movies based on Language '
sns.barplot(x='Language_Id', y='Popularity', data=fmovies_popcount,ax= ax, palette='rainbow', ci=100)
ax.set_xlabel('Language')
ax.set_ylabel('Popularity')
ax.set_title('Popularity of Foreign Movies based on Language')
#ax.invert_yaxis()
```
